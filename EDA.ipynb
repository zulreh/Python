#### **Variáveis**:

**customerID:** ID de cliente<br/>
**gender:** Género (Masculino - male | Feminino - female)<br/>
**SeniorCitizen:** Se o cliente é um cidadão sénior (1, 0)<br/>
**Partner:** Se o cliente tem um(a) parceiro(a) (Yes, No)<br/>
**Dependents:** Se o cliente tem dependentes (Yes, No)<br/>
**tenure:** Número de meses desde que o cliente "entrou" na empresa<br/>
**PhoneService:** Se o cliente tem serviço de telefone (Yes, No)<br/>
**MultipleLines:** Se o cliente tem linhas múltiplas (Yes, No, No phone service)<br/>
**InternetService:** Tipo de ligação à internet (DSL, Fiber optic, No)<br/>
**OnlineSecurity:** Se o cliente contratou o serviço de segurança na internet (Yes, No, No internet service)<br/>
**OnlineBackup:** Se o cliente contratou o serviço de cópia de segurança *onilne* (Yes, No, No internet service)<br/>
**DeviceProtection:** Se o cliente contratou o serviço de protecção de dispositivos (Yes, No, No internet service)<br/>
**TechSupport:** Se o cliente contratou o serviço de suporte técnico (Yes, No, No internet service)<br/>
**StreamingTV:** Se o cliente contratou o serviço de *streaming* de TV (Yes, No, No internet service)<br/>
**StreamingMovies:** Se o cliente contratou o serviço de *streaming* de filmes (Yes, No, No internet service)<br/>
**Contract:** Duração do contrato (Month-to-month, One year, Two year)<br/>
**PaperlessBilling:** Se o cliente contratou o serviço de factura electrónica (Yes, No)<br/>
**PaymentMethod:** Método de pagamento (Electronic check, Mailed check, Bank transfer (automatic), Credit card (automatic))<br/>
**MonthlyCharges:** Cobrança mensal ao cliente<br/>
**TotalCharges:** Cobrança total ao cliente<br/>
**Churn:** Se o cliente deixou a empresa (Yes or No)

#### **Q7:**
Das variáveis numéricas, qual(is) parece(m) ajudar melhor a identificar os clientes que abandonam? (palavra-chave: *pairplot*)

#### **Código:**

```python
cust_data['Churn'].value_counts()
```

```python
sns.set_style('whitegrid')
sns.pairplot(cust_data,hue="Churn",height = 5);
```

```python
#Analisando relação entre MontlyCharges e Tenure para Churn == Yes GRAFICO 2

#Se monthly é >80 e tenure < 20 (range que parece ter mais churn == Yes) 
tmp = cust_data.loc[(cust_data['MonthlyCharges']> 80)&(cust_data['tenure']< 20)&(cust_data['Churn']=='Yes'),
                     ['customerID','tenure','MonthlyCharges']]
percentage1=len(tmp.index)/len(cust_data.loc[(cust_data['MonthlyCharges']> 80)&(cust_data['tenure']< 20)])*100


#Analisando relação entre MontlyCharges e TotalCharges para Churn == Yes GRAFICO 6

#Se monthly é >80, total <2000 e churn == Yes 
tmp1 = cust_data.loc[(cust_data['MonthlyCharges']> 80)&(cust_data['TotalCharges']< 2000)&(cust_data['Churn']=='Yes'),
                     ['customerID','MonthlyCharges','TotalCharges','tenure']]
percentage2=len(tmp1.index)/len(cust_data.loc[(cust_data['MonthlyCharges']> 80)&(cust_data['TotalCharges']< 2000)])*100


#isto é a percentagem de churn == Yes por amostra de clientes totais no range que parece ter mais churn == Yes no grafico 2 e 6
percentage1,percentage2
#e verifica-se que é melhor indicador de abandono a combinação MonthlyCharges com tenure do que a MontlhyCharges com TotalCharges

```

```python
#Analisando se só o facto de ser cliente à pouco tempo tem influencia no abandono
tmp = cust_data.loc[(cust_data['tenure']< 20)&(cust_data['Churn']=='Yes'),
                     ['customerID','tenure']]
percentage=len(tmp.index)/len(cust_data.loc[(cust_data['tenure']< 20)])*100

#Se monthly alto , total <2000 significa que é cliente à pouco tempo tenure entre 1 e 23
tmp1 = cust_data.loc[(cust_data['MonthlyCharges']> 90)&(cust_data['TotalCharges']< 2000)&(cust_data['Churn']== 'Yes'),
                     ['customerID','tenure','MonthlyCharges','TotalCharges']]
tmp1.sort_values('tenure',ascending= False)

#verifica-se que o facto de ser cliente à pouco tempo tem alguma relevância no abandono
percentage
```

```python
#Analisando se só o facto cliente pagar muito por mes tem influencia no abandono

#Se monthly é >80 churn == Yes 
tmp = cust_data.loc[(cust_data['MonthlyCharges']> 80)&(cust_data['Churn']=='Yes'),
                     ['customerID','MonthlyCharges']]
percentage=len(tmp.index)/len(cust_data.loc[(cust_data['MonthlyCharges']> 80)])*100
percentage
```

```python
#Analisando se só o facto cliente pagar muito no total tem influencia no abandono

#Se total é >2000 churn == Yes 
tmp = cust_data.loc[(cust_data['TotalCharges']> 2000)&(cust_data['Churn']=='Yes'),
                     ['customerID','TotalCharges']]
percentage=len(tmp.index)/len(cust_data.loc[(cust_data['TotalCharges']> 2000)])*100
percentage
```

#### **Resposta:**

MonthlyCharges e o tenure, ou seja se o cliente contratou um serviço à menos meses e estiver a pagar mensalmente um valor alto a probalidade de abandono do serviço(churn) é maior (gráfico 2). <br>
Tambem a relação entre TotalCharges e MonthlyCharges, há medida que MonthtlyCharges aumenta e o TotalCharges continua baixo há mais churn (gráfico 6), mas é devido ao tempo de contracto (tenure) que é baixo. <br>
Ou seja a combinação MonthlyCharges e tenure é a que melhor indentifica o abandono do cliente mas individualmente tenure é a variável numérica que melhor a identifica o abandono.

#### **Q8:**

Em relação às variáveis categóricas, quais terão maior potencial preditivo? Analise cuidadosamente os resultados.<br/>
(**Sugestão:** Para cada variável categórica, calcule a % de abandonos em cada categoria - preferencialmente com representação gráfica)

#### **Código:**

```python
categorical=['gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod']
```

```python
for i in categorical:
        tmp4 = cust_data.loc[(cust_data['Churn']=='Yes'),[i,'Churn']]
        tmp = tmp4.groupby(i)['Churn'].count()
        if(i=='SeniorCitizen'):
            tmp = tmp.rename(index={'0': 'No','1' : 'Yes'})
        tmp.plot.pie(autopct='%1.1f%%', startangle=90)
        plt.ylabel(None)
        plt.title('Abandono de clientes por '+i, loc='center', fontdict=dict(fontweight='bold'))
        plt.show()
```

#### **Resposta:**

'SeniorCitizen', 'Dependents', 'PhoneService', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'Contract', 'PaperlessBilling', 'PaymentMethod'. Assumi que as categorias mais preditivas são as categorias em que existe uma variável que é pelo menos superior a 2x as outras variáveis.
