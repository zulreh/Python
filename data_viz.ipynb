````python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
cust_data = pd.read_csv(os.path.join('telco', 'WA_Fn-UseC_-Telco-Customer-Churn.csv') , skipinitialspace=True)
cust_data['TotalCharges'] = cust_data['TotalCharges'].fillna(0)
````

````python
cust_data.describe()
#max tenure 72
#min tenure 0
````

````python
t = cust_data.groupby(['gender'])['customerID'].count()
ax = t.plot.barh(color=['pink','skyblue'])
ax.bar_label(ax.containers[0])


plt.title('Number of clients by Gender', loc='center',fontdict=dict(fontweight='bold'))
sns.despine(bottom=True, left=True)
plt.tick_params(axis='both', bottom=False, left=False, labelbottom=False)
plt.grid(False)
plt.xlabel(None)
plt.ylabel(None)
plt.show()
````

````python
72/12
#ou seja tem clientes que têm contracto no maximo durante 6 anos
````

````python
    
colors= ['greenyellow','yellowgreen','palegreen','darkseagreen','darkolivegreen','green']
tmp4 = cust_data.loc[(cust_data['Contract']=='Month-to-month'),['Contract','tenure']]
cust_data['month_segment'] = pd.cut(cust_data['tenure'], bins=6,precision=0, labels=['0-1 year contract','1-2 year contract','3-4 year contract','4-5 year contract','5-6 year contract','+6 year contract'])
cust_data.groupby('month_segment')['Contract'].count().plot.barh(color=colors);
plt.ylabel(None)
plt.title('Nº of clients by contract year', loc='center', fontdict=dict(fontweight='bold'))
plt.tick_params(axis='y', which='major', left=False, bottom = False, labelbottom = False)
sns.despine(left=True)
plt.show()
````

````python
tmp = cust_data.groupby('tenure')['tenure'].count()
tmp.plot()

plt.xlim(min(cust_data['tenure']), max(cust_data['tenure']))
plt.ylim(min(tmp), max(tmp))

plt.ylabel('Nº Clients')
plt.xlabel('Months of contract')
plt.title('Variation of nº of clients by contract months', loc='center', fontdict=dict(fontweight='bold'))
plt.tick_params(axis='y', which='major', bottom = False, labelbottom = False)
sns.set_style('whitegrid')
plt.show()
````

````python
#quantos clientes por mes de contracto fizeram certo tipo de contrato
grouped_cust_data = cust_data.groupby(['tenure','Contract'])['customerID'].count().to_frame()
grouped_cust_data = grouped_cust_data.reset_index()

sns.scatterplot(x='tenure', y='customerID', data = grouped_cust_data, hue='Contract')
plt.ylabel('Nº Clients')
plt.xlabel('Nº contract months')
plt.title('Nº clients by nº of contract months and the type of contract', loc='center', fontdict=dict(fontweight='bold'))
plt.show()
````

````python
#totalamount de cliente por tipo de contracto
tmp_df = cust_data.groupby(['Contract'])['TotalCharges'].sum().to_frame()
tmp_df =tmp_df.reset_index()
tmp_df = tmp_df.sort_values(by='TotalCharges', ascending = False)

colors= ['blue','green','gold']
fig, ax = plt.subplots()
ax.bar(x=tmp_df['Contract'], height=tmp_df['TotalCharges'],color=colors)

ax.bar_label(ax.containers[0], label_type='edge')
sns.despine(bottom=True, left=True)
plt.tick_params(axis='y', bottom=False, left=False, labelbottom=False, labelleft=False)
plt.grid(False)
plt.title("Total profit by contract type", fontdict=dict(fontweight='bold'))
plt.show()


````

````python
#pie plot de % de utilizadores por tipo de contrato
tmp = cust_data.groupby('Contract')['customerID'].count()
tmp.plot.pie(autopct='%1.1f%%', startangle=90)
plt.ylabel(None)
plt.title('% Clients by contract type', loc='center', fontdict=dict(fontweight='bold'))
plt.show()
````

````python
#paymountMethod com totalamount
tmp_df = cust_data.groupby(['PaymentMethod'])['TotalCharges'].sum().to_frame()
tmp_df = tmp_df.reset_index()
tmp_df = tmp_df.sort_values(by='TotalCharges', ascending = False)

colors=['darkred','gold','lightgray','c']
fig, ax = plt.subplots(figsize=(9, 6))
ax.bar(x=tmp_df['PaymentMethod'], height=tmp_df['TotalCharges'], color=colors)
ax.bar_label(ax.containers[0], label_type='edge')

sns.despine(bottom=True, left=True)
plt.tick_params(axis='y', bottom=False, left=False, labelbottom=False, labelleft=False)
plt.grid(False)
plt.title("Total profit by payment type", fontdict=dict(fontweight='bold'))
plt.show()
````
